```python
# Plot scatter plot of Temperature vs. Weekly Sales
plt.figure(figsize=(10, 6))
plt.scatter(df['Temperature'], df['Weekly_Sales'], alpha=0.5, color='orange')
plt.title('Temperature vs. Weekly Sales')
plt.xlabel('Temperature in Fahrenheit')
plt.ylabel('Weekly Sales')
plt.grid(True) 
plt.show()
```

```python
# Calculate the average weekly sales for each store
avg_sales_per_store = df.groupby('Store')['Weekly_Sales'].mean()

# Plot the bar chart
plt.figure(figsize=(12, 8))
avg_sales_per_store.plot(kind='bar', color='skyblue')
plt.title('Average Weekly Sales by Store')
plt.xlabel('Store')
plt.ylabel('Average Weekly Sales')
plt.xticks(rotation=45)  
plt.grid(axis='y')  
plt.tight_layout() 
plt.show()
```

```python
# Selecting only the numeric columns
numeric_columns = df[['Weekly_Sales', 'Holiday_Flag', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']]

# Calculate the correlation matrix
correlation_matrix = numeric_columns.corr()

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numeric Columns')
plt.show()
```
